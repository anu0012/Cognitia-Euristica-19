```python
features = [x for x in train.columns if x not in ['Avg_damage']]
target = train['Avg_damage']
```

```python
model = GradientBoostingRegressor(max_depth=6,learning_rate=0.09,subsample=0.9,min_impurity_decrease=0.02)
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score, make_scorer
print(cross_val_score(model, train[features], target, cv=5, scoring=make_scorer(r2_score)).mean())
```

```python
model = xgb.XGBRegressor(n_estimators=900, max_depth=8, subsample=1.0, colsample_bytree=0.8)
model.fit(train[features], target)
pred1 = model.predict(test[features])

model = GradientBoostingRegressor()
model.fit(train[features], target)
pred2 = model.predict(test[features])
```

```python
pred = (pred1 * 0.8 + pred2 * 0.2)
```

```python
## make submission
sub = pd.DataFrame()
sub['Avg_damage'] = pred
sub.to_csv('result.csv', index=False,header=False)
```
